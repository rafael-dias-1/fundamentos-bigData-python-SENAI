# Exercício 02 – Unindo dados com Pandas
- a) Crie um dataframe para cada um dos arquivos nba_2015_a.csv, nba_2015_b.csv, nba_2015_c.csv e bust_nba_2015.csv. Chame esses dataframes de df_a, df_b, df_c, bust, respectivamente.
- b) Visualize o head() de cada um dos dataframes.
- c) Concatene os arquivos df_a, df_b e df_c, usando a função concat() usando os índices. Salve um dataframe chamado df_total.
- d) Faça a concatenação do dataframe df_total com o dataframe bust, utilizando a função merge() e a variável ID.
- e) Busque a documentação das funções concat() e merge () e veja quais são os outros parâmetros que podem ser utilizados.

```python
caminho = 'C:\\Users\\FIC\\Documents\\Rafael-Dias-Python\\arquivos\\nba\\'

arquivo = 'nba_2015_a.csv'
df_a = pd.read_csv(caminho + arquivo)

arquivo = 'nba_2015_b.csv'
df_b = pd.read_csv(caminho + arquivo)

arquivo = 'nba_2015_c.csv'
df_c = pd.read_csv(caminho + arquivo)

arquivo = 'bust_nba_2015.csv'
bust = pd.read_csv(caminho + arquivo)
```

```python
df_a.head()
```

```python
df_b.head()
```

```python
df_c.head()
```

```python
bust.head()
```

```python
df_final = pd.concat([df_a, df_b, df_c],ignore_index=True)
```

```python
df_final
```

Faça a concatenação do dataframe df_total com o dataframe bust, utilizando a função merge() e a variável ID.

```python
df_total = pd.merge(df_final,bust,on='ID',how='left')
```
