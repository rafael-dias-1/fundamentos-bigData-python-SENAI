### Importando as bibliotecas (Matplotlib)

Matplotlib é a principal biblioteca para geração de gráficos em Python. A biblioteca possui uma infinidade de possibilidades, sendo possível criarmos desde gráficos simples até outros mais elaborados.


```python
import matplotlib.pyplot as plt # Permite a geração de gráficos
import numpy as np # Permite trabalhar com matrizes e vetores
```

##### Primeiro Gráfico com Matplotlib

```python
# Criando os vetores (conjunto de dados)
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Plotar o gráfico

plt.plot(x, y)
```

##### Modificando o tipo dos marcadores

```python
plt.plot(x, y, marker= '*')
```

#### Alterando as cores da linha

```python
plt.plot(x, y, marker = '*', c='#DEA643')
```

##### Mudando como os pontos são ligados e o tipo de linha

```python
plt.plot(
    x, y,               #Os dados do gráfico
    marker = '*',       # Tipo de marcador
    c = 'green',        # Cor da linha
    ds = 'steps',       # A forma com que os pontos são conectados
    ls = '--'           # Estilo da linha
)
```

##### Mudando a cor dos pontos

```python
plt.plot (
    x,y,                # Dados do gráfico
    marker = '*',       # Marcadores
    c = 'red',          # Cor da linha
    ds = 'steps',       # Forma de conexão dos pontos
    ls = '--',          # Estilo da linha
    mec = 'blue'        # Cor dos marcadores
)
```

#### Matplotlib – adicionando outros elementos

Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outras possibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando a atenção do que realmente importa.

```python
# Adiciona dados ao gráfico
plt.plot(x,y)

# Adicionar título ao eixo X
plt.xlabel('Valores de X', fontsize = 12)

# Adicionar título ao eixo Y
plt.ylabel('Valores de Y', fontsize = 12)

# Adicionar título ao gráfico
plt.title ('Título do gráfico', fontsize = 18, c='purple')

# Altera as cores dos valores dos eixos

plt.xticks(fontsize = 8, c ='blue')

plt.yticks(fontsize = 8, c ='green')
```

#### Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
z = np.cos(x) 
plt.plot (x,y) 
plt.plot (x,z)
```

#### Matplotlib – Múltiplos subgráficos

```python
# O comando plt.subplot permite criar grupo de gráfico
# plt.subplot (linhas, colunas, Posição elemento)

plt.subplot(2,1,1) # Primeiro gráfico
plt.plot(x,y) # Conjunto de dados do primeiro gráfico
plt.title('subplots') 

plt.subplot(2,1,2) # Segundo gráfico
plt.plot(x,z) # Conjunto de dados do segundo gráfico
```

### Gráficos na vertical

```python
plt.subplot(1,2,1) 
plt.plot(x,y) 
plt.title('subplots') 
plt.subplot(1,2,2) 
plt.plot(x,z)
```

## Seaborn - Gráficos estatístivos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# carregar o dataset iris da biblioteca SEABORN

iris = sns.load_dataset('iris')

iris.head()
```

```python
# criando um histograma para análise univariada
# kde é a densidade (curva de probabilidade)
sns.distplot(iris['petal_length'], kde = True)
```

```python
#criando a curva de densidade

sns.distplot(iris['petal_length'], hist = False)
```

Seaborn - Pairplot (bivariada)

```python
sns.pairplot(
    iris, # dataset
    hue= 'species', # legenda
    diag_kind= 'kde', # tipo de grafico de distribuição
    kind=  'scatter', # tipo de grafico de correlação
    palette='husl' # Conjunto de cores
)
plt.savefig('pairplot_iris.png')
```
